```python
class Node:
    def __init__(self, data):
        
        self.nodo_pai = None
        self.direita = None
        self.esquerda = None
        self.cima = None
        self.baixo = None
        self.data = data
        self.h = 0
        self.f = 0
        self.g = 0
        
    def insert(self, data):
        if self.direita is None:
            self.direita = Node(data)            
        elif self.esquerda is None:
            self.esquerda = Node(data)
        elif self.cima is None:
            self.cima = Node(data)
        elif self.baixo is None:
            self.baixo = Node(data)
```

```python
def busca_valor(m, x):
    i= 0
    j= 0
    while (i<size) and (m[i,j]!= x):
        while (j<size) and (m[i,j]!= x):
            j=j+1
        if j!=size and m[i,j]==x:
            break
        else:
            j=0
            i=i+1
    return i,j
```

```python
def heuristica(m):
    h = 0
    for i in range(0, size-1):
        for j in range(0, size-1):
            if m[i,j] != saida_esperada[i,j]:
                a, b = busca_valor(saida_esperada, m[i,j])
                h = abs(i-a) + abs(j-b)            
    return h
```

```python
def prioridade(dado):
    return dado.f

solucao_encontrada = False
nos_abertos = 0

lista_dados = list()
lista_dados.append(Node(m))
matriz_embaralhada = m.copy()
nos_abertos = nos_abertos +1

def abre_no(raiz): 

    rs, v = movimento(raiz)
    for i in range (0,v):
        rs[i].nodo_pai = raiz
        rs[i].h = heuristica(rs[i].data)
        rs[i].g = raiz.f - raiz.h + 1
        rs[i].f = rs[i].g + rs[i].h
        raiz.insert(rs[i].data)        
    lista_dados.extend(rs)  
    lista_dados.sort(key = prioridade)
```
